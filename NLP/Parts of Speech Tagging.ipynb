Part-of-Speech (POS) Tagging is the process of assigning a grammatical category (such as noun, verb, adjective, adverb, etc.) to each word in a sentence based on its meaning and context.

```python
paragraph = """A. P. J. Abdul Kalam was a renowned scientist and the former President of India.
He inspired millions of students through his speeches, books, and vision for the country's development.
Dr. Kalam believed that education and hard work are the keys to success. He encouraged young people to dream big,
think creatively, and contribute to society with dedication and integrity."""
```

```python
from nltk.tokenize import sent_tokenize
from nltk.tokenize import word_tokenize
from nltk.corpus import stopwords
```

```python
sentences = sent_tokenize(paragraph)
```

```python
from nltk import pos_tag
```

```python
import nltk
nltk.download('averaged_perceptron_tagger_eng')
```

```python
for i in range(len(sentences)):
    words = word_tokenize(sentences[i])
    words = [word for word in words if word not in stopwords.words('english')]
    pos_tag1 += pos_tag(words) 
    #sentences[i] = ' '.join(words)
```
