#### Lemmatization is a text preprocessing technique in NLP that converts a word to its base or dictionary form (lemma) by considering the word's meaning and grammatical context.

Examples:

    -running → run
    -better → good
    -studies → study
    -am, is, are → be

WordNetLemmatizer is a lemmatization tool provided by NLTK that uses the WordNet lexical database to convert words into their dictionary (base) form called a lemma.

```python
from nltk.stem import WordNetLemmatizer
```

```python
lemmatizer = WordNetLemmatizer()
```

```python
import nltk
nltk.download('wordnet')
```

```python
lemmatizer.lemmatize('better', pos = 'a')
```

```python
lemmatizer.lemmatize('raining', pos = 'v')
```

```python
lemmatizer.lemmatize('writing', pos = 'v')
```

```python
lemmatizer.lemmatize('Preprocess', pos = 'v')
```
