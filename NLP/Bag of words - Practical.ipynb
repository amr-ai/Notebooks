```python
import pandas as pd
```

```python
messages = pd.read_csv(r"SMSSpamCollection", sep = '\t', names = ['label', 'message'], encoding = 'utf-8')
```

```python
import re
import nltk
nltk.download('stopwords')
```

```python
from nltk.tokenize import sent_tokenize
from nltk.tokenize import word_tokenize
from nltk.corpus import stopwords
from nltk.stem import PorterStemmer
from nltk.stem import SnowballStemmer
from nltk.stem import RegexpStemmer
```

```python
ps = PorterStemmer()
```

```python
## Cleaning the punctuationns, full stop, commas etc from the data.
## Converting the all words to lower case
## Splitting the sentences to make the list (which is useful in stemming)
## Stemming the words using Porter Stemmer
## Joining the words again to make final sentences
## Appending this in corpus varible

corpus  = []

for i in range(0, len(messages)):
    review = re.sub('[^a-zA-Z]',' ', messages['message'][i])
    review = review.lower()
    review = review.split()
    review = [ps.stem(word) for word in review if word not in stopwords.words('english')]
    review = ' '.join(review)
    corpus.append(review)
```
